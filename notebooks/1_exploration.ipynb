Cell 3: Data Inspection & Visualization

```python
try:
    # Initialize the dataset pointing to your local 'data' folder
    # Expected structure: DL-VIS_Github/data/train/images/
    ds_exploration = PanNukeDataset(root_dir=DATA_DIR, split='train', transform=True)
    print(f"✅ Success! Found {len(ds_exploration)} images in the local training set.")

    # Retrieve a sample for visual verification
    img, mask = ds_exploration[0]

    # Professional Visualization
    fig, ax = plt.subplots(1, 2, figsize=(12, 6))
    
    # Image processing for display (Convert from Tensor and handle normalization)
    img_display = img.permute(1, 2, 0).numpy()
    if img_display.min() < 0: # Check if the image was normalized (mean/std)
        img_display = img_display * 0.2 + 0.5 # Approximate inverse normalization
        
    ax[0].imshow(img_display.clip(0, 1))
    ax[0].set_title("Histopathology Image (RGB)")
    ax[0].axis('off')

    # 'jet' colormap is useful for highlighting different cell types in the mask
    ax[1].imshow(mask.numpy(), cmap='jet')
    ax[1].set_title("Segmentation Mask (Ground Truth)")
    ax[1].axis('off')
    
    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"❌ Error during data loading: {e}")
    print(f"🔍 Technical Check: Ensure your local path exists: {os.path.join(DATA_DIR, 'train', 'images')}")
```
